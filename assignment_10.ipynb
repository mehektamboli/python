```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv("archive (4) (1).csv")


sales_col = "Sales"


plt.figure(figsize=(6, 4))
plt.boxplot(df[sales_col].dropna(), vert=False, patch_artist=True)


plt.title(f"Boxplot of {sales_col}")
plt.xlabel(sales_col)
plt.grid(axis='x', linestyle='--', alpha=0.7)

plt.show()
```
