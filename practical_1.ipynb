```python
import numpy as np

A = np.array([[4, 2],
              [3, 5]])

rank = np.linalg.matrix_rank(A)

determinant = np.linalg.det(A)

transpose = np.transpose(A)

inverse = np.linalg.inv(A)

eigenvalues, eigenvectors = np.linalg.eig(A)

print("Matrix A:\n", A)
print("\nRank of A:", rank)
print("Determinant of A:", determinant)
print("Transpose of A:\n", transpose)
print("Inverse of A:\n", inverse)
print("Eigenvalues:", eigenvalues)
print("Eigenvectors:\n", eigenvectors)

arr = np.array([5, 2, 9, 1, 7, 3])

max_element = np.max(arr)
min_element = np.min(arr)

print("Array:", arr)
print("Maximum element:", max_element)
print("Minimum element:", min_element)

sorted_arr = np.sort(arr)
print("Sorted array:", sorted_arr)

mean_value = np.mean(arr)
print("Mean of array:", mean_value)

A = np.array([[1, 2],
              [3, 4]])
B = np.array([[5, 6],
              [7, 8]])


result = np.dot(A, B)

print("Matrix A:\n", A)
print("Matrix B:\n", B)
print("Product of A and B:\n", result)
```
