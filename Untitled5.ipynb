```python
import pandas as pd
import numpy as np


batsman_A = [46, 34, 52, 78, 65, 81, 26, 46, 19, 47]
batsman_B = [59, 25, 81, 47, 73, 78, 42, 35, 42, 10]


df = pd.DataFrame({
    "Batsman A": batsman_A,
    "Batsman B": batsman_B
})


stats = pd.DataFrame({
    "Mean": [np.mean(batsman_A), np.mean(batsman_B)],
    "Variance": [np.var(batsman_A, ddof=1), np.var(batsman_B, ddof=1)],
    "Standard Deviation": [np.std(batsman_A, ddof=1), np.std(batsman_B, ddof=1)],
    "Coefficient of Variation": [
        np.std(batsman_A, ddof=1)/np.mean(batsman_A),
        np.std(batsman_B, ddof=1)/np.mean(batsman_B)
    ]
}, index=["Batsman A", "Batsman B"])

print("Descriptive Statistics:\n", stats)


if stats.loc["Batsman A", "Mean"] > stats.loc["Batsman B", "Mean"]:
    print("\nBatsman A has a higher average score.")
else:
    print("\nBatsman B has a higher average score.")

if stats.loc["Batsman A", "Coefficient of Variation"] < stats.loc["Batsman B", "Coefficient of Variation"]:
    print("Batsman A is more consistent.")
else:
    print("Batsman B is more consistent.")
```
