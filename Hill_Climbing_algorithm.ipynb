```python
import copy

def heuristic(state, goal):
    score = 0
    goal_stack = goal[0]

    for stack in state:
        for i, block in enumerate(stack):
            if i < len(goal_stack) and block == goal_stack[i]:
                score += 1
    return score



def generate_neighbors(state):
    neighbors = []

    for i in range(len(state)):
        if not state[i]:
            continue


        new_state = copy.deepcopy(state)
        block = new_state[i].pop()


        for j in range(len(state)):
            if i != j:
                temp_state = copy.deepcopy(new_state)
                temp_state[j].append(block)
                neighbors.append(temp_state)


        temp_state = copy.deepcopy(new_state)
        temp_state.append([block])
        neighbors.append(temp_state)

    return neighbors



def hill_climbing(initial, goal):
    current = initial
    current_score = heuristic(current, goal)

    while True:
        neighbors = generate_neighbors(current)
        best_neighbor = None
        best_score = current_score

        for neighbor in neighbors:
            score = heuristic(neighbor, goal)
            if score > best_score:
                best_score = score
                best_neighbor = neighbor

        if best_neighbor is None:
            break

        current = best_neighbor
        current_score = best_score

    return current



initial_state = [['P', 'Q', 'S', 'T'], ['R']]
goal_state = [['P', 'Q', 'R','S', 'T']]

solution = hill_climbing(initial_state, goal_state)

print("Initial State:", initial_state)
print("Goal State:", goal_state)
print("Final State after Hill Climbing:", solution)
```
