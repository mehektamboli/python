```python
from pgmpy.models import DiscreteBayesianNetwork
from pgmpy.factors.discrete import TabularCPD
from pgmpy.inference import VariableElimination
model = DiscreteBayesianNetwork([
    ('StudyHours', 'ExamPerformance'),
    ('Intelligence', 'ExamPerformance'),
    ('Sleep', 'ExamPerformance'),
    ('ExamPerformance', 'Grade')
])

cpd_study = TabularCPD(
    variable='StudyHours',
    variable_card=2,
    values=[[0.3],
            [0.7]]
)


cpd_intelligence = TabularCPD(
    variable='Intelligence',
    variable_card=2,
    values=[[0.4],
            [0.6]]
)


cpd_sleep = TabularCPD(
    variable='Sleep',
    variable_card=2,
    values=[[0.2],
            [0.8]]
)


cpd_exam = TabularCPD(
    variable='ExamPerformance',
    variable_card=2,
    values=[
        [0.9, 0.6, 0.5, 0.2, 0.6, 0.3, 0.2, 0.1],
        [0.1, 0.4, 0.5, 0.8, 0.4, 0.7, 0.8, 0.9]
    ],
    evidence=['StudyHours', 'Intelligence', 'Sleep'],
    evidence_card=[2, 2, 2]
)
cpd_grade = TabularCPD(
    variable='Grade',
    variable_card=2,
    values=[
        [0.5, 0.5],
        [0.5, 0.5]
    ],
    evidence=['ExamPerformance'],
    evidence_card=[2]
)


model.add_cpds(cpd_study, cpd_intelligence, cpd_sleep, cpd_exam, cpd_grade)

print("Model is valid:", model.check_model())


inference = VariableElimination(model)

result = inference.query(
    variables=['Grade'],
    evidence={'StudyHours': 1, 'Sleep': 1}
)

print("\nProbability of Grade given StudyHours=High and Sleep=Good:")
print(result)
```
