```python
import pandas as pd
data = {
    'Name': ['Amit', 'Sneha', 'Ravi', 'Priya', 'Karan'],
    'Age': [23, 21, 25, 22, 24],
    'Marks': [85, 90, 78, 92, 88],
    'City': ['Pune', 'Mumbai', 'Delhi', 'Chennai', 'Kolkata']
}

df = pd.DataFrame(data)

print("Original DataFrame:\n", df)

df.to_csv('students.csv', index=False)
print("\nData exported to 'students.csv'")



print("\nFirst 3 Rows:\n", df.head(3))


print("\nInformation about dataset:")
print(df.info())

print("\nStatistical Summary:\n", df.describe())

print("\nMaximum Marks:", df['Marks'].max())
print("Minimum Marks:", df['Marks'].min())

sorted_df = df.sort_values(by='Marks', ascending=False)
print("\nData sorted by Marks:\n", sorted_df)

high_scorers = df[df['Marks'] > 85]
print("\nStudents with marks > 85:\n", high_scorers)


df['Grade'] = ['A', 'A+', 'B', 'A+', 'A']
print("\nDataFrame after adding Grade column:\n", df)


new_df = pd.read_csv('students.csv')
print("\nData imported from 'students.csv':\n", new_df)
```
