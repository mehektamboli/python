```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

np.random.seed()
days = pd.date_range(start="2024-12-01", periods=30)
print(days)
```

```python
data = {
    "Date": np.tile(days, 4),
    "City": np.repeat(["Pune", "Gujarat", "Rajasthan", "Hyderabad"], 30),
    "Temperature": np.concatenate([
        np.random.normal(36, 2.5, 30),   # Pune
        np.random.normal(28, 1.8, 30), # Gujarat
        np.random.normal(24, 1.9, 30), # Rajasthan
        np.random.normal(36, 1.2, 30)  # Hyderabad
    ])
}
df = pd.DataFrame(data)
print(df)
```

```python

stats_summary = df.groupby("City")["Temperature"].agg(
    Mean="mean",
    Median="median",
    Mode=lambda x: stats.mode(x, keepdims=False).mode
).reset_index()

print(" Comparative Temperature Statistics:\n")
print(stats_summary)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.subplot(1,2,1)
sns.histplot(data=df, x="Temperature", hue="City", kde=False, bins=15)
plt.title("Temperature Distribution Across Cities")
```

```python

plt.subplot(1,2,2)
sns.boxplot(x="City", y="Temperature", data=df, palette="deep")
plt.title("Temperature Spread Across Cities")
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(data=df, x="Date", y="Temperature", hue="City", marker="o")
plt.title("Daily Temperature Variation Across Cities")
plt.xticks(rotation=45)
plt.show()
```
